```python
# The dataset contains 4 header rows with logo, so we skip them using header=4
# The first column is empty, so we remove it using iloc

df = pd.read_excel('/content/Adidas US Sales Datasets.xlsx', header=4)
df = df.iloc[:, 1:]
```

```python
# Preview the dataset after removing extra header rows and the empty first column

print("Shape of the dataset:", df.shape)
print("First 5 rows:")
df.head()
```

```python
#Basic dataset information

print("\nColumn names:")
print(df.columns)
```

```python
print("\nData types and non-null counts:")
df.info()
```

```python
# Separate numerical and categorical columns

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print("Numerical Columns:")
print(numerical_cols)

print("\nCategorical Columns:")
print(categorical_cols)
```

```python
print("\nSummary statistics for numerical columns:")
df.describe()
```

```python
#Check for missing values

print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Check for duplicate rows

duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")
```

```python
# Verify data types

print(df.dtypes)
```

```python
# Dataset is already clean: no missing values, no duplicates, and correct data types.
```

```python
# Extract Month and Year from 'Invoice Date'

df['Month'] = df['Invoice Date'].dt.month_name()
df['Year'] = df['Invoice Date'].dt.year
```

```python
# Preview the dataset after adding derived columns

print("Shape of dataset after adding derived columns:", df.shape)
df.head()
```

```python
# Aggregate total sales by Year

sales_by_year = df.groupby('Year')['Total Sales'].sum().reset_index()
print("Total Sales by Year:")
display(sales_by_year)
```
